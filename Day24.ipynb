# Applying Function

```python
import pandas as pd
def salary_raise(x):
    return x+2
```

```python
df = pd.DataFrame({'Employee ID':[111, 222, 333, 444],
                   'Employee Name':['Chanel', 'Steve', 'Mitch', 'Bird'],
                   'Salary [$/h]':[35, 29, 38, 20],
                   'Years of Experience':[3, 4 ,9, 1]})
df
```

```python
df['new_salary']=df['Salary [$/h]'].apply(salary_raise)
```

```python
df
```

```python
df["emplen"]=df["Employee Name"].apply(len)
df
```

```python
df[['Employee ID','emplen','Salary [$/h]','new_salary','Years of Experience']]
```

```python
def check(y):
    if y.islower():
        return "yes"
    else:
        return "no"
```

```python
df["check"]=df["Employee Name"].apply(check)
df
```

```python
df.sort_values(by='Years of Experience',ascending=False,inplace=True)
```

```python
df[df["Salary [$/h]"]>20]
```

# Concatinating and merging



Reference: https://pandas.pydata.org/pandas-docs/stable/merging.html

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
```

```python
pd.concat(([df1,df2,df3]),axis=0)
```

```python
df1=pd.concat(([df1,df2,df3]),axis=1)
```

```python
df1.style.set_caption("Concat And Merging")
```
