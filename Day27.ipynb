```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Plot

```python
x=np.arange(0,10,0.2)#evenly sampled time at 0.2s interval
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Sin Wave")
plt.title("first plot")
plt.show()
```

```python
z=np.tan(x)
plt.plot(x,z)
plt.xlabel("Time")
plt.ylabel("cos Wave")
plt.title("2 plot")
plt.show()
```

```python
plt.figure(figsize=(10,10))
X=np.random.randn(1000)
Y=np.random.randn(1000)
plt.scatter(X,Y)
plt.xlabel("time")
plt.ylabel("sinwave")
plt.title("1st scatter plot")
```

```python
t=np.arange(0,5,0.2)
plt.plot(t,t,"r",t,t**2,"b",t,t**3,"g^",t,t**4,"bs",linewidth=8.0)
```
